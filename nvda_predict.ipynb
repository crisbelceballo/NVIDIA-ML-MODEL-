```python
import yfinance as yf  #libreria que permite ingresar a los datos financieros de Yahoo Finance
```

```python
nvda = yf.Ticker('NVDA') #Obtener la informacion detallada de NIVDIA
```

```python
nvda = nvda.history(period='max') #Obtener el mayor numero de historial de precios de NVIDIA
```

```python
nvda 
```

```python
nvda.index
```

```python
nvda.plot.line(y='Close', use_index = True) #Generar un gráfico de los datos de cierre de las acciones de NVIDIA
```

```python
nvda['Tomorrow'] = nvda['Close'].shift(-1) #crear columna que contenga la fecha de cierre del dia siguiente
```

```python
nvda
```

```python
nvda['Target'] = (nvda['Tomorrow'] > nvda['Close']).astype(int) #Determinar si el cierre futuro es mayor que el cierre actual
```

```python
nvda
```

```python
nvda = nvda.loc["2000-01-01":].copy() #Seleccionar datos desde el 1 de enero del 2000
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators= 100, min_samples_split=100, random_state= 1)

train = nvda.iloc[:-100]
test = nvda.iloc[-100:]

predictors = ['Close', 'Volume', 'Open', 'High', 'Low']
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score

predictions = model.predict(test[predictors])
```

```python
import pandas as pd 
predictions = pd.Series(predictions, index = test.index)
```

```python
precision_score(test['Target'], predictions)
```

```python
combined = pd.concat([test['Target'], predictions], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors,model): 
     model.fit(train[predictors], train['Target'])
     predictions = model.predict(test[predictors])
     predictions = pd.Series(predictions, index= test.index, name='Predictions' )
     combined = pd.concat([test['Target'], predictions], axis= 1)
     return combined
```

```python
def backtest(data, model, predictors, start= 2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step): 
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(nvda, model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
predictions['Target'].value_counts() / predictions.shape[0]
```

```python
horizons = [2, 5, 60, 250, 1000]

new_predictors = []

for horizon in horizons:
    rolling_averages = nvda.rolling(horizon).mean()
    
    ratio_column = f'Close Ratio_{horizon}'
    nvda[ratio_column] = nvda['Close'] / rolling_averages['Close']
    
    trend_column = f'Trend_{horizon}'
    nvda[trend_column] = nvda.shift(1).rolling(horizon).sum()['Target']
    
    new_predictors += [ratio_column, trend_column]
```

```python
nvda =nvda.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict (train, test, predictors, model): 
          model.fit(train[predictors], train['Target'])
          predictions = model.predict_proba(test[predictors])[:,1]
          predictions[predictions >= .6] = 1
          predictions[predictions < .6] = 0 
          predictions = pd.Series(predictions, index=test.index, name ='Predictions')
          combined = pd.concat([test['Target'],predictions], axis = 1)
          return combined
```

```python
predictions = backtest(nvda, model, new_predictors)
```
